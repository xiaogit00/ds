# Intro to Matplotlib

```python
import matplotlib.pyplot as plt # API for adding plot elements, stores state of object, can plot continuously
import matplotlib
```

```python
fig, ax = plt.subplots() #creates a figure and subplot
plt.style.use('ggplot')

# What's a figure? -> container for everything you want to draw. can contain multiple subplots, and other elements like titles, legends

# Ax -> axes -> where the actual plotting happens (draw lines, bars, etc)
```

```python
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
fig
```

#### By default, ax.plot(x, y) will take an array of x and y values

```python
fig2, axs2 = plt.subplots(2, 2)  # 2 rows, 2 columns
axs2[0, 0].plot([1, 2, 3], [1, 4, 9])
axs2[0, 1].plot([1, 2, 3], [1, 2, 3])
axs2[1, 0].plot([1, 2, 3], [1, 0, 1])
axs2[1, 1].plot([1, 2, 3], [3, 2, 1])

# creates 2 by 2 plots 
```

```python
axs2[0, 0].set_title("Figure 1")
```

```python
fig2
```

```python
fig3, ax3 = plt.subplots(nrows=2, ncols=1, figsize=(8, 6), sharex=True)
ax3[0].plot([1, 2, 3], [1, 2, 3])
ax3[1].plot([1, 2, 3], [3, 2, 1])
```
